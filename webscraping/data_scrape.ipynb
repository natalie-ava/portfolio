## Yelp Functions

```python
## Yelp Functions

def url_to_soup(url):
  response = requests.get(url)
  time.sleep(11)
  if response.status_code != 200:
    print('Failed')
  soup = BeautifulSoup(response.text,'html.parser')
  return soup

# Restaurant Names
def get_restaurant_names(soup):
  titles = soup.find_all('a',class_='css-19v1rkv')
  return [item.text for item in titles]

# Restaurant Reviews
def get_reviews(soup):
    number_of_reviews = soup.find_all('span', class_='css-chan6m')
    restaurant_reviews = []

    for item in number_of_reviews:
        text = item.text
        if 'reviews' in text:
            restaurant_reviews.append(item.text)

    return restaurant_reviews

# Restaurant Stars
def get_number_of_stars(soup):
    star_count = soup.find_all('span',class_="css-gutk1c")
    restaurant_stars = []

    for item in star_count:
        st = item.text
        if re.search('0|1|2|3|4|5',st):
            restaurant_stars.append(item.text)

    return restaurant_stars

# Restaurant Price Range
def get_price_range(soup):
  results1 = soup.find_all('span',class_='priceRange__09f24__ZgJXy css-blvn7s')
  results2 = soup.find_all('p',class_='priceRange__09f24__ZgJXy css-chan6m')
  price_range = results1 + results2
  return [item.text for item in price_range]

# Restaurant Short Links
def get_short_links(soup):
  restaurant_links = soup.find_all('h3')
  return[item.a.get('href') for item in restaurant_links]
```

## Math Jobs Functions

```python
#math job functions

def url_to_soup(url):
  response = requests.get(url)
  if response.status_code != 200:
    print('Failed')
  soup = BeautifulSoup(response.text,'html.parser')
  return soup

# Institution name
def get_institution(soup):
  location = soup.find_all('b')
  return[item.text for item in location]

# Deadline
def get_deadline(soup):
    results1 = soup.find_all('span', class_='sml')
    results2 = soup.find_all('span', class_='purplesml')

    deadlines = []
    for item in results1:
        text = item.text.strip()
        if 'PM' in text:
            deadlines.append(text)

    for item in results2:
        text = item.text.strip()
        if 'PM' in text:
            deadlines.append(text)
    try:
      d_line = deadlines[0].replace('deadline','').replace('(','').replace(')','').strip()
    except:
      d_line = 'na'
    return d_line
```
