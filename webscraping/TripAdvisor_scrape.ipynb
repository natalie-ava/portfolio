## All Functions

```python
import time
import requests
from bs4 import BeautifulSoup

def url_to_soup(url):
  response = requests.get(url)
  time.sleep(11)
  if response.status_code != 200:
    print('Failed')
  soup = BeautifulSoup(response.text,'html.parser')
  return soup

#Restaurant Name
def get_restaurant_names(soup):
  titles = soup.find_all('div',class_='biGQs _P fiohW alXOW NwcxK GzNcM ytVPx UTQMg RnEEZ ngXxk')
  return [item.text.split('. ', 1)[-1] for item in titles]

# Restaurant Price Range
def get_price_range(soup):
  price_range = soup.find_all('span',class_="YECgr")
  restaurant_price = []
  for item in price_range:
    text = item.text
    if '$' in text:
        restaurant_price.append(text)
  return restaurant_price

# Restaurant Bubble Count
def get_bubbles(soup):
  bubble_count = soup.find_all('span', class_='Qqwyj')
  number_of_bubbles = []
  for item in bubble_count:
    text = item.text
    if 'bubbles' in text:
      number_of_bubbles.append(text)
  return number_of_bubbles

# Restaurant Links
def get_restaurant_links(soup):
    def format_restaurant_name(name):
        no_accents = ''.join(
            char for char in unicodedata.normalize('NFD', name)
            if unicodedata.category(char) != 'Mn'
        )
        return re.sub(r'\W+', '_', no_accents)
    titles = soup.find_all('div', class_='biGQs _P fiohW alXOW NwcxK GzNcM ytVPx UTQMg RnEEZ ngXxk')
    names = [item.text.split('. ', 1)[-1] for item in titles]
    urls = []
    for name in names:
        formatted_name = format_restaurant_name(name)
        url = f'https://www.tripadvisor.com/Restaurant_Review-g60956-d26336409-Reviews-{formatted_name}-San_Antonio_Texas.html'
        urls.append(url)
    return urls

get_restaurant_links(soup)

# Restaurant Type
def get_restaurant_type(soup):
    restaurant_type = soup.find_all('span', class_="YECgr Tsrjt")
    types = []
    for item in restaurant_type:
        text = item.text.strip()
        if ',' in text and not any(char.isdigit() for char in text):
            types.append(text)
    return types

get_restaurant_type(soup)

# Restaurant Review Count

def get_number_of_reviews(soup):
    review_count = soup.find_all('span', class_="YECgr Tsrjt")
    review_counts = []
    for item in review_count:
        text = item.text.strip()
        match = re.search(r'(\d+)\s+reviews?', text)
        if match:
            review_count = match.group(1)
            review_counts.append(review_count)
    return review_counts

get_number_of_reviews(soup)
```

```python
def get_restaurant_data(soup):
    names = get_restaurant_names(soup)
    price_range = get_price_range(soup)
    bubbles = get_bubbles(soup)
    urls = get_restaurant_links(soup)
    cuisines = get_restaurant_type(soup)
    reviews = get_number_of_reviews(soup)

    max_length = max(len(names), len(price_range), len(bubbles), len(urls), len(cuisines), len(reviews))

    names += [None] * (max_length - len(names))
    price_range += [None] * (max_length - len(price_range))
    bubbles += [None] * (max_length - len(bubbles))
    urls += [None] * (max_length - len(urls))
    cuisines += [None] * (max_length - len(cuisines))
    reviews += [None] * (max_length - len(reviews))

    data = {
        'restaurant name': names,
        'price range': price_range,
        'bubbles (rating)': bubbles,
        'Link': urls,
        'type of food': cuisines,
        'number of reviews': reviews
    }

    df = pd.DataFrame(data)
    return df

get_restaurant_data(soup)
```
